```python
import numpy as np
import pylab as pl
from sklearn import linear_model, preprocessing
from sklearn.datasets import load_svmlight_file
import sys

def do(file_src, y_limit, ylabel, figure_index):
    X_train, y_train = load_svmlight_file(file_src)

    X_train = X_train.toarray()

    pl.figure(num=figure_index+1, figsize=(10, 6))
    pl.scatter(X_train[:,0], y_train, s=3, color='black')
    pl.xlabel('Avg Bitrate/kbps')
    pl.ylabel(ylabel)
    pl.xlim((0, 2300))
    pl.ylim((0, y_limit))
    
    pl.figure(num=figure_index+2, figsize=(10, 6))
    pl.scatter(X_train[:,1], y_train, s=3, color='black')
    pl.xlabel('Switch times')
    pl.ylabel(ylabel)
    pl.xlim((0, 3000))
    pl.ylim((0, y_limit))
    
    pl.figure(num=figure_index+3, figsize=(10, 6))
    pl.scatter(X_train[:,2], y_train, s=3, color='black')
    pl.xlabel('Platform')
    pl.ylabel(ylabel)
    pl.xticks([1, 2, 3, 4, 5, 6, 7, 8],
              ['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown'])
    pl.ylim((0, y_limit))
    
    pl.figure(num=figure_index+4, figsize=(10, 6))
    pl.scatter(X_train[:,3], y_train, s=3, color='black')
    pl.xlabel('Initial timestamp')
    pl.ylabel(ylabel)
    pl.xlim((0, 24))
    pl.ylim((0, y_limit))
```

```python
do('./data-archive/time_train4', 23000, 'User Watching Lasting Time/s', 0)
```

```python
do('./data-archive/ad_train4', 1, 'Advertisement Ratio', 10)
```
