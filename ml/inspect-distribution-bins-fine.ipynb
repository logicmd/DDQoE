```python
'''
Platform:  
{ 
     iPad: 10264251 
     Android: 7512890 
     iPhone: 5377486 
     Apple TV: 368609 
     iPod: 243399 
     Mac: 21074 
     Windows: 423 
     None: 133 
}

[1, 2, 3, 4, 5, 6, 7, 8],
['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown']
'''
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file

platform = { 
                   'All': -1,
                   'Apple TV': 1, 
                   'Mac': 2, 
                   'Windows': 3, 
                   'iPad': 4, 
                   'Android': 5, 
                   'iPhone': 6, 
                   'iPod' : 7, 
                   'None' :8, 
                }
        
metric = {
                  'Watching Lasting Time(s)': 0,
                  'Ad Ratio': 1,
                  'Avg Bitrate(kbps)': 2,
                  'Switch(per hour)': 3,
                  'Initial Time(\'o clock)': 5
              }
train = None

class cdf:

    def __init__(self):
        global train
        
        X_train, y_train = load_svmlight_file('./data_fine/time_train')
        X_train, y_train2 = load_svmlight_file('./data_fine/ad_train')
        X_train = X_train.toarray()
        train = np.column_stack((y_train, y_train2, X_train))


    def plot(self, m, p, n, b = False, lower = 0, upper = Inf, bins = 5):
        global platform
        global metric
        global train
        
        sample = None
        mask = np.zeros((train.shape[0], ), dtype=bool)
        mask.fill(True)
        if b:
            mask = np.logical_and(train[:, metric[m]] > lower, train[:, metric[m]] < upper) 
        
        
        if platform[p] < 0:
            sample = train[mask, metric[m]]
        else:
            mask2 = np.logical_and(mask, train[:,4]==platform[p])
            sample = train[mask2, metric[m]]
        
        print 'Sample size: ' + str(sample.shape[0])
        
        percentile = np.linspace(0, 1, bins)
        print percentile
        
        
        if sample.shape[0] > 0:
            ecdf = sm.distributions.ECDF(sample)
            plt.figure(num=n, figsize=(10, 6))
            plt.plot(ecdf.x, ecdf.y)
            
            xbins = []
            index = []
            for pos in xrange(len(percentile)):
                st = 0
                if len(index) > 0:
                    st = index[-1]

                
                for ind in xrange(st, len(ecdf.y)):
                    if ecdf.y[ind] >= percentile[pos]:
                        x_val = ecdf.x[ind]
                        if x_val == -inf:
                            x_val = ecdf.x[ind+1]
                        xbins.append(x_val)
                        index.append(ind)
                        break
                        
            print xbins
            
            plt.xlabel(p + ": " + m)
            plt.ylabel('CDF')
            plt.show()




```

```python
c = cdf()
d = 'iPad'
c.plot('Watching Lasting Time(s)', d, 1, bins = 3)
c.plot('Ad Ratio', d, 1, bins = 3)
c.plot('Avg Bitrate(kbps)', d, 1, bins = 3)
c.plot('Switch(per hour)', d, 1, bins = 3)
c.plot('Initial Time(\'o clock)', d, 1, bins = 3)
```
