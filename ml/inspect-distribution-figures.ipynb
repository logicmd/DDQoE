```python
'''
Platform:  
{ 
     iPad: 10264251 
     Android: 7512890 
     iPhone: 5377486 
     Apple TV: 368609 
     iPod: 243399 
     Mac: 21074 
     Windows: 423 
     None: 133 
}

[1, 2, 3, 4, 5, 6, 7, 8],
['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown']
'''
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file

platform = { 
                   'All': -1,
                   'Apple TV': 1, 
                   'Mac': 2, 
                   'Windows': 3, 
                   'iPad': 4, 
                   'Android': 5, 
                   'iPhone': 6, 
                   'iPod' : 7, 
                   'None' :8, 
                }
        
metric = {
                  'Watching Lasting Time(s)': 0,
                  'Ad Ratio': 1,
                  'Avg Bitrate(kbps)': 2,
                  'Switch(per hour)': 3,
                  'Initial Time(\'o clock)': 5
              }
train = None

class cdf:

    def __init__(self):
        global train
        
        X_train, y_train = load_svmlight_file('./data/time_train4')
        X_train, y_train2 = load_svmlight_file('./data/ad_train4')
        X_train = X_train.toarray()
        train = np.column_stack((y_train, y_train2, X_train))


    def plot(self, m, p, n, b = False, lower = 0, upper = Inf, multiple = False, name=''):
        global platform
        global metric
        global train
        
        sample = None
        mask = np.zeros((train.shape[0], ), dtype=bool)
        mask.fill(True)
        if b:
            mask = np.logical_and(train[:, metric[m]] > lower, train[:, metric[m]] < upper) 
        
        
        

        
        samples = []
        
        if multiple:
            for device in ['iPad', 'iPhone', 'Android', 'Apple TV']:
                cur_mask = np.logical_and(mask, train[:,4]==platform[device])
                samples.append(train[cur_mask, metric[m]])

        else:
            if platform[p] >= 0:
                mask = np.logical_and(mask, train[:,4]==platform[p])

            samples.append(train[mask, metric[m]])
        
        
        
        if len(samples) > 0 and samples[0].shape[0] > 0:
            plt.figure(num=n, figsize=(5.6, 3.36))
            
            if not multiple:
                sample = samples[0]
                ecdf = sm.distributions.ECDF(sample)
                pp, = plt.plot(ecdf.x, ecdf.y, linewidth=2, c='k')
                plt.legend([pp], ['All Devices'], loc=4)
                p = 'All Devices'
            else:
                plots = []
                for sample in samples:
                    if len(sample) == 0:
                        continue
                        
                    ecdf = sm.distributions.ECDF(sample)
                    plot, = plt.plot(ecdf.x, ecdf.y, linewidth=2)

                    plots.append(plot)
                
                text = ['iPad', 'iPhone', 'Android']
                if m != 'Ad Ratio':
                    text.append('Apple TV')
                
                plt.legend(plots, text, loc=4)
                
                p = 'Various Devices'
            
            
            plt.xlabel(p + ": " + m)
            plt.ylabel('CDF')
            #plt.show()
            plt.savefig('/home/tangkai/DDQoE/ml/figure/' + name + '.eps', format='eps')
            




```

```python
c = cdf()
fig_num = 1
for m in metric.keys():
    c.plot(m, 'All', fig_num, multiple = False, name='all_ad')
    c.plot(m, '', fig_num + 1, multiple = True, name='v_ad')
    fig_num += 2
```
