```python
'''
Platform:  
{ 
     iPad: 10264251 
     Android: 7512890 
     iPhone: 5377486 
     Apple TV: 368609 
     iPod: 243399 
     Mac: 21074 
     Windows: 423 
     None: 133 
}

[1, 2, 3, 4, 5, 6, 7, 8],
['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown']
'''
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file

platform = { 
                   'All': -1,
                   'Apple TV': 1, 
                   'Mac': 2, 
                   'Windows': 3, 
                   'iPad': 4, 
                   'Android': 5, 
                   'iPhone': 6, 
                   'iPod' : 7, 
                   'None' :8, 
                }
        
metric = {
                  'Watching Lasting Time(s)': 0,
                  'Ad Ratio': 1,
                  'Avg Bitrate(kbps)': 2,
                  'Switch(per hour)': 3,
                  'Initial Time(\'o clock)': 5
              }
train = None

class cdf:

    def __init__(self):
        global train
        
        X_train, y_train = load_svmlight_file('./data/time_train4')
        X_train, y_train2 = load_svmlight_file('./data/ad_train4')
        X_train = X_train.toarray()
        train = np.column_stack((y_train, y_train2, X_train))


    def plot(self, m, p, n, b = False, lower = 0, upper = Inf):
        global platform
        global metric
        global train
        
        sample = None
        mask = np.zeros((train.shape[0], ), dtype=bool)
        mask.fill(True)
        if b:
            mask = np.logical_and(train[:, metric[m]] > lower, train[:, metric[m]] < upper) 
        
        
        if platform[p] < 0:
            sample = train[mask, metric[m]]
        else:
            mask2 = np.logical_and(mask, train[:,4]==platform[p])
            sample = train[mask2, metric[m]]
        
        print 'Sample size: ' + str(sample.shape[0])
        
        if sample.shape[0] > 0:
            ecdf = sm.distributions.ECDF(sample)
            
            x = np.linspace(min(sample), max(sample))
            y = ecdf(x)
            plt.figure(num=n, figsize=(10, 6))
            plt.step(x, y)
            plt.xlabel(p + ": " + m)
            plt.ylabel('CDF')
            plt.show()




```

```python
c = cdf()
m = 'Watching Lasting Time(s)'
c.plot(m, 'All', 0, True, 0)
c.plot(m, 'iPad', 1, True, 0 )
c.plot(m, 'Android', 2, True, 0 )
c.plot(m, 'iPhone', 3, True, 0 )
c.plot(m, 'Apple TV', 4, True, 0)
c.plot(m, 'Mac', 5, True, 0)
c.plot(m, 'Windows', 6, True, 0)
```
