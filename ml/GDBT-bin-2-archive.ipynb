```python
'''
Platform:  
{ 
     iPad: 10264251 
     Android: 7512890 
     iPhone: 5377486 
     Apple TV: 368609 
     iPod: 243399 
     Mac: 21074 
     Windows: 423 
     None: 133 
}

[1, 2, 3, 4, 5, 6, 7, 8],
['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown']
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import cross_val_score
from sklearn.datasets import load_svmlight_file
from sklearn.cross_validation import train_test_split, KFold
from sklearn.ensemble import GradientBoostingClassifier


from sklearn import tree
from six import StringIO  
import os

platform = { 
                   'All': -1,
                   'Apple TV': 1, 
                   'Mac': 2, 
                   'Windows': 3, 
                   'iPad': 4, 
                   'Android': 5, 
                   'iPhone': 6, 
                   'iPod' : 7, 
                   'None' :8, 
                }

metric_list = ['Watching Lasting Time(s)',
               'Ad Ratio',
               'Avg Bitrate(kbps)',
               'Switch(per hour)',
               'Initial Time(\'o clock)']

metric = {
                  'Watching Lasting Time(s)': 0,
                  'Ad Ratio': 1,
                  'Avg Bitrate(kbps)': 2,
                  'Switch(per hour)': 3,
                  'Initial Time(\'o clock)': 5
              }
train = None

bins = {
            'iPad': {
                'Watching Lasting Time(s)': [0, 10000, 300000],
                'Ad Ratio': [0, 0.7, 1.1],
                'Avg Bitrate(kbps)': [0, 2000, 2500],
                'Switch(per hour)': [0, 0.12, 61],
                'Initial Time(\'o clock)': [0, 15, 24]
            },
        }
        

class gdbt:

    def __init__(self):
        global train
        
        X_train, y_train = load_svmlight_file('./archive/data2/time_train4')
        X_train, y_train2 = load_svmlight_file('./archive/data2/ad_train4')
        X_train = X_train.toarray()
        train = np.column_stack((y_train, y_train2, X_train))


    def train_and_test(self, feature, label, n=10, test_size=0.1):
        scores = []
        max_score = -1
        max_clf = None
        
        kf = KFold(n=label.shape[0], n_folds=5)
        for train_index, test_index in kf:
            feature_train, feature_test = feature[train_index], feature[test_index]
            label_train, label_test = label[train_index], label[test_index]

            clf = GradientBoostingClassifier()
            clf = clf.fit(feature_train, label_train)
            score = clf.score(feature_test, label_test)
            if score > max_score:
                max_score = score
                max_clf = clf
                train_score = clf.train_score_
            scores.append(score)
        print "scores: " + str(scores)
        return max_score, max_clf
            
    def plot(self, p):
        global platform
        global metric
        global train
        
        
        mat = np.zeros((train[train[:,4]==platform[p],:].shape[0],5))
        for idx, m in enumerate(metric_list):

            mat[:, idx] = numpy.digitize(train[train[:,4]==platform[p], metric[m]], bins[p][m])

        
        print 'Sample size: ' + str(mat.shape[0])
        
        time_train = mat[:, 0]
        ad_train = mat[:, 1]
        features = mat[:, 2:5]
        
        
        time_score, time_clf = self.train_and_test(features, time_train)
        print "Best scores:" + str(time_score)
        print "Importance:" + str(time_clf.feature_importances_)
        print "Score of each decisiton tree" + str(time_clf.train_score_)
        
        ad_score, ad_clf = self.train_and_test(features, ad_train)
        print "Best scores:" + str(ad_score)
        print "Importance:" + str(ad_clf.feature_importances_)
        print "Score of each decisiton tree" + str(ad_clf.train_score_)
        
    
        
        
        

if __name__ == '__main__':
    c = gdbt()
    c.plot('iPad')
```
