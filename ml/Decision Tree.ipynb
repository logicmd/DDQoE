```python
'''
Platform:  
{ 
     iPad: 10264251 
     Android: 7512890 
     iPhone: 5377486 
     Apple TV: 368609 
     iPod: 243399 
     Mac: 21074 
     Windows: 423 
     None: 133 
}

[1, 2, 3, 4, 5, 6, 7, 8],
['Apple TV', 'Mac', 'Windows', 'iPad', 'Android', 'iPhone', 'iPod', 'Unknown']
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_svmlight_file

from sklearn import tree
from six import StringIO  
import os

platform = { 
                   'All': -1,
                   'Apple TV': 1, 
                   'Mac': 2, 
                   'Windows': 3, 
                   'iPad': 4, 
                   'Android': 5, 
                   'iPhone': 6, 
                   'iPod' : 7, 
                   'None' :8, 
                }

metric_list = ['Watching Lasting Time(s)',
               'Ad Ratio',
               'Avg Bitrate(kbps)',
               'Switch(per hour)',
               'Initial Time(\'o clock)']

metric = {
                  'Watching Lasting Time(s)': 0,
                  'Ad Ratio': 1,
                  'Avg Bitrate(kbps)': 2,
                  'Switch(per hour)': 3,
                  'Initial Time(\'o clock)': 5
              }
train = None

bins = {
            'iPad': {
                'Watching Lasting Time(s)': [0, 10000, 300000],
                'Ad Ratio': [0, 0.7, 1.1],
                'Avg Bitrate(kbps)': [0, 2000, 2500],
                'Switch(per hour)': [0, 0.12, 60],
                'Initial Time(\'o clock)': [0, 15, 24]
            },
        }
        

class cdf:

    def __init__(self):
        global train
        
        X_train, y_train = load_svmlight_file('./data/time_train4')
        X_train, y_train2 = load_svmlight_file('./data/ad_train4')
        X_train = X_train.toarray()
        train = np.column_stack((y_train, y_train2, X_train))

    def output(self, file_name, clf):
        with open(file_name+".dot", 'w') as f:
            f = tree.export_graphviz(clf, out_file=f)
            with open('./decision_tree/convert.sh', 'wa') as ff:
                ff.write("dot -Tpdf " + file_name +".dot" + " -o " + file_name + ".pdf")
            ff.close()


    def plot(self, p, n):
        global platform
        global metric
        global train
        
        
        mat = np.zeros((train[train[:,4]==platform[p],:].shape[0],5))
        for idx, m in enumerate(metric_list):
            #print idx, m
            mat[:, idx] = numpy.digitize(train[train[:,4]==platform[p], metric[m]], bins[p][m])
            #print train[:, metric[m]]
            #print mat[:, idx]
        
        print 'Sample size: ' + str(mat.shape[0])
        
        time_train = mat[:, 0]
        ad_train = mat[:, 1]
        features = mat[:, 2:5]
        
        time_clf = tree.DecisionTreeClassifier()
        time_clf = time_clf.fit(features, time_train)
        time_file_name = "./decision_tree/"+ p + "_time"
            
        ad_clf = tree.DecisionTreeClassifier()
        ad_clf = ad_clf.fit(features, ad_train)
        ad_file_name = "./decision_tree/"+ p + "_ad"
        
        with open(time_file_name+".dot", 'w') as f:
            f = tree.export_graphviz(time_clf, out_file=f)

            os.unlink(f)
            f.close()
            
        with open(ad_file_name+".dot", 'w') as f:
            f = tree.export_graphviz(ad_clf, out_file=f)
            
            os.unlink(f)
            f.close()
            
        with open('./decision_tree/convert.sh', 'w') as ff:
            ff.write("#!/bin/sh\n")
            ff.write("dot -Tpdf " + time_file_name +".dot" + " -o " + time_file_name + ".pdf\n")
            ff.write("dot -Tpdf " + ad_file_name +".dot" + " -o " + ad_file_name + ".pdf\n") 
        ff.close()
        

if __name__ == '__main__':
    c = cdf()
    c.plot('iPad', 1)

```
